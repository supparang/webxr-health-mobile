# S3 Coding Lab — Breadth-First Search

BFS ใช้ Queue สำรวจโหนดตามระดับและเหมาะกับการหาเส้นทางสั้นในกราฟที่ไม่มีน้ำหนัก


## Predict
ลำดับ visited จะเป็นอย่างไร

เขียนคำตอบก่อนรันโค้ด

```python
from collections import deque

graph = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['E'],
    'D': [],
    'E': []
}

queue = deque(['A'])
visited = []

while queue:
    node = queue.popleft()
    if node not in visited:
        visited.append(node)
        queue.extend(graph[node])

print(visited)
```
